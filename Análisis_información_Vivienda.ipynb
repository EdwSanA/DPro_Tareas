```python
!pip install -q missingno seaborn

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from scipy import stats

# Mostrar gráficos inline
%matplotlib inline
sns.set_context("notebook")

# ===== 1) Localizar y cargar el archivo =====
# Busca 'train.csv' en el directorio actual
cwd_files = os.listdir()
csv_candidates = [f for f in cwd_files if f.lower().endswith('.csv')]
if not csv_candidates:
    raise FileNotFoundError(
        "No se encontró ningún archivo .csv en el directorio actual. "
        "Debes subir el archivo 'train.csv' descargado desde Kaggle."
    )

# Prefiere 'train.csv' si existe
csv_name = 'train.csv' if 'train.csv' in csv_candidates else csv_candidates[0]
print(f"Usando archivo: {csv_name}")

df = pd.read_csv(csv_name)
print("\n== Dimensiones del dataset ==")
print(df.shape)

# ===== 2) Inspección inicial =====
print("\n== Primeras 5 filas ==")
display(df.head())

print("\n== Tipos de datos por columna ==")
display(pd.DataFrame(df.dtypes, columns=['dtype']).T)

print("\n== Descripción estadística (numéricas) ==")
display(df.describe().T)

# Confirmar variable objetivo
target_col = "SalePrice"
if target_col not in df.columns:
    raise KeyError(f"No se encontró la variable objetivo '{target_col}' en el dataset.")
print(f"\nLa variable objetivo es: {target_col}")

# ===== 3) Valores faltantes =====
missing_counts = df.isnull().sum().sort_values(ascending=False)
missing_ratio = (missing_counts / len(df)).sort_values(ascending=False)
missing_df = pd.DataFrame({
    'TotalMissing': missing_counts,
    'MissingRatio': missing_ratio
})
print("\n== Valores faltantes por columna (top 30) ==")
display(missing_df[missing_df.TotalMissing > 0].head(30))

# Visualizar con missingno
plt.figure(figsize=(10,4))
msno.bar(df, fontsize=8)
plt.show()

plt.figure(figsize=(10,4))
msno.matrix(df, fontsize=8)
plt.show()

plt.figure(figsize=(10,6))
msno.heatmap(df)
plt.show()

# Eliminar columnas con >=5 valores faltantes
cols_to_drop = missing_counts[missing_counts >= 5].index.tolist()
print(f"\nColumnas con >=5 valores faltantes (se eliminarán):")
print(cols_to_drop)

df_reduced = df.drop(columns=cols_to_drop)
print("\nDimensiones después de eliminar columnas con muchos missing:")
print(df_reduced.shape)

# Eliminar filas con valores faltantes restantes
rows_before = len(df_reduced)
df_reduced = df_reduced.dropna(axis=0)
rows_after = len(df_reduced)
print(f"Filas antes de dropna: {rows_before}; filas después: {rows_after}; "
      f"filas eliminadas: {rows_before-rows_after}")

# ===== 4) Distribución de la variable objetivo =====
y = df[target_col].dropna()
print("\n== Estadísticas de SalePrice (original) ==")
print(y.describe())

skew_orig = stats.skew(y)
kurt_orig = stats.kurtosis(y)
print(f"Asimetría (skew) original: {skew_orig:.4f}")
print(f"Curtosis (exceso) original: {kurt_orig:.4f}")

plt.figure(figsize=(10,4))
sns.histplot(y, kde=True)
plt.title("Distribución de SalePrice (original)")
plt.xlabel("SalePrice")
plt.show()

# Transformación logarítmica
y_log = np.log1p(y)
skew_log = stats.skew(y_log)
kurt_log = stats.kurtosis(y_log)
print(f"Asimetría (skew) log-transformada: {skew_log:.4f}")
print(f"Curtosis (exceso) log-transformada: {kurt_log:.4f}")

plt.figure(figsize=(10,4))
sns.histplot(y_log, kde=True)
plt.title("Distribución de SalePrice (log1p transformada)")
plt.xlabel("log1p(SalePrice)")
plt.show()

print("""
Interpretación:
- La distribución original suele mostrar sesgo (skew) positivo y alta curtosis.
- La transformación logarítmica la acerca a una distribución normal,
  lo que puede mejorar el rendimiento de modelos de regresión.
""")
```
