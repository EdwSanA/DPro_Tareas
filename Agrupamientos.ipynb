```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
import pandas as pd
import seaborn as sns

# Configuracin para Google Colab
plt.style.use('default')
sns.set_palette("husl")

"""

 IMPLEMENTACIN DE K-MEANS DESDE CERO


Propsito del Sprint:
- Comprender el algoritmo K-means a travs de implementacin desde cero
- Aprender tcnicas de clustering no jerrquico
- Implementar mtodos de evaluacin de clusters
- Aplicar clustering a datos reales

Qu es K-means?
K-means es un algoritmo de clustering no jerrquico que agrupa datos en K clusters.
El objetivo es minimizar la suma de errores al cuadrado (SSE) entre los puntos de datos
y sus centroides asignados.

Algoritmo K-means:
1. Inicializar K centroides aleatoriamente
2. Asignar cada punto al centroide ms cercano
3. Recalcular centroides como promedio de puntos asignados
4. Repetir pasos 2-3 hasta convergencia

Mtricas de evaluacin:
- SSE (Sum of Squared Errors): Minimizar
- Silhouette Score: Maximizar (rango -1 a 1)
- Mtodo del codo: Encontrar punto de inflexin en SSE vs K

================================================================================
"""

class ScratchKMeans():
    """
    K-means clustering scratch implementation

    Parameters
    ----------
    n_clusters : int
        Number of clusters
    n_init : int
        Number of times to run with different initial centroids
    max_iter : int
        Maximum number of iterations
    tol : float
        Tolerance for convergence
    verbose : bool
        True to output the learning process
    """

    def __init__(self, n_clusters=3, n_init=10, max_iter=300, tol=1e-4, verbose=False):
        # Record hyperparameters as attributes
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.max_iter = max_iter
        self.tol = tol
        self.verbose = verbose
        self.centroids = None
        self.labels = None
        self.sse = None
        self.is_fitted = False

    """

    PROBLEMA 1: Determinar el valor inicial del punto central


    Cree un cdigo que determine el valor inicial de _K a partir de K puntos centrales
    _1 a partir de los datos de entrada. K es el nmero especificado de clsteres.

    La forma ms sencilla de determinar el valor inicial es seleccionar aleatoriamente
    K de los X_n puntos de datos. Por favor, implemente esto esta vez.

    El resultado del aprendizaje de K-medias se ve afectado por el valor inicial del
    punto central . Por lo tanto, durante el aprendizaje, el clculo se realiza con
    mltiples valores iniciales.
    ================================================================================
    """
    def _initialize_centroids(self, X):
        """
        Initialize centroids by randomly selecting K data points

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Input data

        Returns
        -------
        ndarray, shape (n_clusters, n_features)
            Initial centroids
        """
        n_samples, n_features = X.shape
        # Randomly select K data points as initial centroids
        random_indices = np.random.choice(n_samples, self.n_clusters, replace=False)
        centroids = X[random_indices].copy()

        if self.verbose:
            print(f"Initialized {self.n_clusters} centroids randomly")

        return centroids

    """

    PROBLEMA 2: Creacin de una funcin para obtener SSE


    Crea una funcin que calcule la suma de errores al cuadrado (SSE).

    El propsito de K-means es encontrar r_nk y _k que minimicen este SSE.
    A partir de los resultados de agrupamiento calculados con mltiples valores
    iniciales, utilice la funcin para encontrar este SSE al decidir cul adoptar finalmente.

    SSE = (n=1 to N) (k=1 to K) r_nk * ||X_n - _k||

    Donde:
    - n: ndice de puntos de datos
    - k: ndice de clster
    - X_n: n-simo punto de datos
    - _k: k-simo punto central
    - r_nk: 1 si el punto de datos X_n pertenece al grupo k, 0 en caso contrario
    ================================================================================
    """
    def _calculate_sse(self, X, centroids, labels):
        """
        Calculate Sum of Squared Errors (SSE)

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Input data
        centroids : ndarray, shape (n_clusters, n_features)
            Cluster centroids
        labels : ndarray, shape (n_samples,)
            Cluster labels for each data point

        Returns
        -------
        float
            Sum of Squared Errors
        """
        sse = 0.0
        for k in range(self.n_clusters):
            # Get points assigned to cluster k
            cluster_points = X[labels == k]
            if len(cluster_points) > 0:
                # Calculate squared distances to centroid
                distances = np.sum((cluster_points - centroids[k]) ** 2, axis=1)
                sse += np.sum(distances)

        return sse

    """
    PROBLEMA 3: Asignacin al clster
    Escriba el cdigo que asigne todos los puntos de datos X_n al punto central ms cercano _k.

    Cuanto ms cerca est el k-meno, ms se reduce la distancia euclidiana entre los puntos.
    La distancia euclidiana se calcula mediante el teorema de Pitgoras, pero puede expresarse
    mediante la siguiente frmula para los vectores p y q:

    ||q - p|| = ((q - p)  (q - p))

    Utilice esta funcin tal como se proporciona np.linalg.norm.

    Esto significa que el punto central _k es fijo y el SSE se minimiza, pero no es necesario
    encontrar el SSE en s.
    ================================================================================
    """
    def _assign_clusters(self, X, centroids):
        """
        Assign each data point to the nearest centroid

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Input data
        centroids : ndarray, shape (n_clusters, n_features)
            Cluster centroids

        Returns
        -------
        ndarray, shape (n_samples,)
            Cluster labels for each data point
        """
        n_samples = X.shape[0]
        labels = np.zeros(n_samples, dtype=int)

        for i in range(n_samples):
            # Calculate distances to all centroids
            distances = np.linalg.norm(X[i] - centroids, axis=1)
            # Assign to nearest centroid
            labels[i] = np.argmin(distances)

        return labels

    """

    PROBLEMA 4: Movimiento del punto central


    Escriba el cdigo que mueve el punto central _k a la media (centroide) de todos
    los puntos de datos X_n asignados al k-simo grupo.

    Ha corregido la asignacin de clster r_nk y ha minimizado el SSE, pero no necesita
    solicitar el SSE en s.
    ================================================================================
    """
    def _update_centroids(self, X, labels):
        """
        Update centroids to be the mean of assigned points

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Input data
        labels : ndarray, shape (n_samples,)
            Cluster labels for each data point

        Returns
        -------
        ndarray, shape (n_clusters, n_features)
            Updated centroids
        """
        centroids = np.zeros((self.n_clusters, X.shape[1]))

        for k in range(self.n_clusters):
            # Get points assigned to cluster k
            cluster_points = X[labels == k]
            if len(cluster_points) > 0:
                # Update centroid to mean of assigned points
                centroids[k] = np.mean(cluster_points, axis=0)
            else:
                # If no points assigned, keep previous centroid
                centroids[k] = self.centroids[k] if self.centroids is not None else np.zeros(X.shape[1])

        return centroids

    """

    PROBLEMA 5: Repetir


    Escriba cdigo que aprenda asignando repetidamente al clster y moviendo el punto central.

    Detenga esto si se cumple alguna de las siguientes condiciones:
    - La repeticin alcanza el nmero especificado de veces. (Argumento max_iter en la plantilla)
    - La suma de la diferencia entre el punto central _k y el centro de gravedad es menor
      o igual a la tolerancia especificada. (Argumento tol en la plantilla)
    ================================================================================
    """
    def _fit_single_init(self, X):
        """
        Run K-means with a single initialization

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Input data

        Returns
        -------
        tuple
            (centroids, labels, sse, n_iterations)
        """
        # Initialize centroids
        centroids = self._initialize_centroids(X)

        for iteration in range(self.max_iter):
            # Store previous centroids for convergence check
            prev_centroids = centroids.copy()

            # Assign points to clusters
            labels = self._assign_clusters(X, centroids)

            # Update centroids
            centroids = self._update_centroids(X, labels)

            # Check for convergence
            centroid_shift = np.linalg.norm(centroids - prev_centroids)

            if self.verbose and iteration % 10 == 0:
                sse = self._calculate_sse(X, centroids, labels)
                print(f"Iteration {iteration}: SSE = {sse:.4f}, Centroid shift = {centroid_shift:.6f}")

            if centroid_shift < self.tol:
                if self.verbose:
                    print(f"Converged at iteration {iteration}")
                break

        # Calculate final SSE
        final_labels = self._assign_clusters(X, centroids)
        sse = self._calculate_sse(X, centroids, final_labels)

        return centroids, final_labels, sse, iteration + 1

    """

    PROBLEMA 6: Calcular con diferentes valores iniciales


    Cree cdigo que calcule la repeticin de la pregunta 5 modificando el valor inicial
    un nmero especfico de veces. El nmero de veces se especifica mediante el argumento
    n_init en la plantilla.

    Dado que el resultado final del aprendizaje ser el que tenga el SSE ms pequeo entre
    mltiples clculos, mantenga el punto central _k en ese momento.
    ================================================================================
    """
    def fit(self, X):
        """
        Calculate clustering by K-means

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Features of training data
        """
        X = np.array(X)

        if self.verbose:
            print(f"Starting K-means with {self.n_clusters} clusters")
            print(f"Running {self.n_init} initializations")

        best_sse = float('inf')
        best_centroids = None
        best_labels = None

        # Run K-means multiple times with different initializations
        for init in range(self.n_init):
            if self.verbose:
                print(f"\nInitialization {init + 1}/{self.n_init}")

            centroids, labels, sse, n_iter = self._fit_single_init(X)

            if self.verbose:
                print(f"SSE: {sse:.4f}, Iterations: {n_iter}")

            # Keep the best result (lowest SSE)
            if sse < best_sse:
                best_sse = sse
                best_centroids = centroids.copy()
                best_labels = labels.copy()

        # Store the best results
        self.centroids = best_centroids
        self.labels = best_labels
        self.sse = best_sse
        self.is_fitted = True

        if self.verbose:
            print(f"\nBest result: SSE = {best_sse:.4f}")

    """

    PROBLEMA 7: Estimacin


    Al estimar a qu grupo pertenecer para los nuevos datos, haga lo mismo que en la
    primera fase de entrenamiento.

    Para el punto de datos X_n y el punto central _k determinados por el aprendizaje,
    seleccione r_nk que minimice SSE.

    En concreto es como sigue:
    - Asigna el punto de datos X_n al _k ms cercano.
    ================================================================================
    """
    def predict(self, X):
        """
        Calculate which cluster the input data belongs to

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Input data

        Returns
        -------
        ndarray, shape (n_samples,)
            Cluster labels
        """
        if not self.is_fitted:
            raise ValueError("Model must be fitted before making predictions")

        X = np.array(X)
        return self._assign_clusters(X, self.centroids)

    def get_sse(self):
        """
        Get the Sum of Squared Errors of the fitted model

        Returns
        -------
        float
            SSE value
        """
        if not self.is_fitted:
            raise ValueError("Model must be fitted before getting SSE")
        return self.sse


"""

PROBLEMA 8: Implementacin del mtodo del codo

En K-medias, el nmero de puntos centrales k debe determinarse artificialmente,
y existe el mtodo del codo para determinarlo. Escriba el cdigo para considerar
el nmero de puntos centrales k mediante el mtodo del codo.

Muestra un grfico lineal con el SSE en el eje vertical y el nmero de clsteres k
en el eje horizontal. Para ello, asegrese de poder obtener el SSE de una instancia de K-means.
"""
def elbow_method(X, max_k=10, verbose=False):
    """
    Implement elbow method to find optimal number of clusters

    Parameters
    ----------
    X : ndarray, shape (n_samples, n_features)
        Input data
    max_k : int
        Maximum number of clusters to test
    verbose : bool
        Whether to print progress

    Returns
    -------
    tuple
        (k_values, sse_values)
    """
    k_values = range(1, max_k + 1)
    sse_values = []

    for k in k_values:
        if verbose:
            print(f"Testing k = {k}")

        kmeans = ScratchKMeans(n_clusters=k, n_init=10, verbose=False)
        kmeans.fit(X)
        sse_values.append(kmeans.get_sse())

    # Plot elbow curve
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, sse_values, 'bo-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Sum of Squared Errors (SSE)')
    plt.title('Elbow Method for Optimal k')
    plt.grid(True)
    plt.show()

    return k_values, sse_values


"""

PROBLEMA 9: (Tarea avanzada) Diagrama de silueta


Tambin hay un diagrama de silueta para ayudar a determinar k. Cree cdigo para mostrar
el diagrama de silueta.

El coeficiente de silueta s_n para el punto de datos n se puede calcular mediante la
siguiente frmula:

s_n = (b_n - a_n) / max(b_n, a_n)

Donde:
- a_n: Cohesin. La distancia promedio a otros puntos de datos en el mismo clster.
- b_n: Grado de divergencia. Distancia promedio a todos los puntos de datos del otro grupo ms cercano.

Idealmente, la divergencia b es grande y la cohesin a es pequea, con un coeficiente
de silueta de s cercano a 1.
"""
def calculate_silhouette_coefficients(X, labels):
    """
    Calculate silhouette coefficients for each data point

    Parameters
    ----------
    X : ndarray, shape (n_samples, n_features)
        Input data
    labels : ndarray, shape (n_samples,)
        Cluster labels

    Returns
    -------
    tuple
        (silhouette_vals, silhouette_avg)
    """
    n_samples = X.shape[0]
    n_clusters = len(np.unique(labels))

    if n_clusters < 2:
        return np.zeros(n_samples), 0.0

    silhouette_vals = np.zeros(n_samples)

    for i in range(n_samples):
        # Calculate a_i (cohesion): average distance to points in same cluster
        same_cluster_points = X[labels == labels[i]]
        if len(same_cluster_points) > 1:
            a_i = np.mean([np.linalg.norm(X[i] - point) for point in same_cluster_points if not np.array_equal(X[i], point)])
        else:
            a_i = 0

        # Calculate b_i (separation): minimum average distance to other clusters
        b_i_values = []
        for cluster_id in np.unique(labels):
            if cluster_id != labels[i]:
                other_cluster_points = X[labels == cluster_id]
                if len(other_cluster_points) > 0:
                    avg_dist = np.mean([np.linalg.norm(X[i] - point) for point in other_cluster_points])
                    b_i_values.append(avg_dist)

        if b_i_values:
            b_i = min(b_i_values)
        else:
            b_i = 0

        # Calculate silhouette coefficient
        if max(a_i, b_i) > 0:
            silhouette_vals[i] = (b_i - a_i) / max(a_i, b_i)
        else:
            silhouette_vals[i] = 0

    silhouette_avg = np.mean(silhouette_vals)
    return silhouette_vals, silhouette_avg


def plot_silhouette_diagram(X, labels, silhouette_vals, silhouette_avg):
    """
    Plot silhouette diagram

    Parameters
    ----------
    X : ndarray, shape (n_samples, n_features)
        Input data
    labels : ndarray, shape (n_samples,)
        Cluster labels
    silhouette_vals : ndarray, shape (n_samples,)
        Silhouette coefficients
    silhouette_avg : float
        Average silhouette coefficient
    """
    from matplotlib import cm

    n_clusters = len(np.unique(labels))
    cluster_labels = np.unique(labels)

    y_ax_lower, y_ax_upper = 0, 0
    yticks = []

    plt.figure(figsize=(10, 8))

    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[labels == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(i / n_clusters)
        plt.barh(
            range(y_ax_lower, y_ax_upper),
            c_silhouette_vals,
            height=1.0,
            edgecolor="none",
            color=color,
        )
        yticks.append((y_ax_lower + y_ax_upper) / 2)
        y_ax_lower += len(c_silhouette_vals)

    plt.axvline(silhouette_avg, color="red", linestyle="--")
    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel("Cluster")
    plt.xlabel("Silhouette coefficient")
    plt.title(f"Silhouette Diagram (Average: {silhouette_avg:.3f})")
    plt.show()


"""

CONJUNTO DE DATOS ARTIFICIALES PARA AGRUPAMIENTO

"""
def create_sample_dataset():
    """Create artificial dataset for clustering experiments"""
    X, _ = make_blobs(
        n_samples=100,
        n_features=2,
        centers=4,
        cluster_std=0.5,
        shuffle=True,
        random_state=0,
    )
    return X


def visualize_clusters(X, labels, centroids=None, title="K-means Clustering"):
    """
    Visualize clustering results

    Parameters
    ----------
    X : ndarray, shape (n_samples, n_features)
        Input data
    labels : ndarray, shape (n_samples,)
        Cluster labels
    centroids : ndarray, shape (n_clusters, n_features), optional
        Cluster centroids
    title : str
        Plot title
    """
    plt.figure(figsize=(10, 8))

    # Plot data points
    scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.7)
    plt.colorbar(scatter)

    # Plot centroids if provided
    if centroids is not None:
        plt.scatter(centroids[:, 0], centroids[:, 1],
                   c='red', marker='x', s=200, linewidths=3, label='Centroids')
        plt.legend()

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.show()


"""

ANLISIS DE DATOS DE MAYORISTAS

"""

def load_wholesale_data():
    """
    Load wholesale customers dataset
    Note: In practice, you would load from a CSV file
    For demonstration, we'll create synthetic data similar to the real dataset
    """
    np.random.seed(42)

    # Create synthetic wholesale data
    n_samples = 440

    # Channel: 1=Horeca (Hotel/Restaurant/Caf), 2=Retail
    channel = np.random.choice([1, 2], n_samples, p=[0.3, 0.7])

    # Region: 1=Lisbon, 2=Oporto, 3=Other
    region = np.random.choice([1, 2, 3], n_samples, p=[0.4, 0.3, 0.3])

    # Create correlated spending data
    data = []
    for i in range(n_samples):
        # Base spending varies by channel and region
        base_spending = 1000 + channel[i] * 500 + region[i] * 200

        # Add some noise
        noise = np.random.normal(0, 200)

        # Create spending categories (correlated)
        fresh = max(0, base_spending * np.random.uniform(0.1, 0.3) + noise)
        milk = max(0, base_spending * np.random.uniform(0.05, 0.15) + noise * 0.5)
        grocery = max(0, base_spending * np.random.uniform(0.2, 0.4) + noise * 0.8)
        frozen = max(0, base_spending * np.random.uniform(0.05, 0.15) + noise * 0.3)
        detergents = max(0, base_spending * np.random.uniform(0.02, 0.08) + noise * 0.2)
        delicatessen = max(0, base_spending * np.random.uniform(0.01, 0.05) + noise * 0.1)

        data.append([channel[i], region[i], fresh, milk, grocery, frozen, detergents, delicatessen])

    columns = ['Channel', 'Region', 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']
    df = pd.DataFrame(data, columns=columns)

    return df


"""

PROBLEMA 10: Seleccin del nmero de conglomerados k


Seleccione la cantidad de clsteres k que considere ptima. Adems de utilizar el
mtodo del codo y el diagrama de silueta, la forma en que los humanos desean usarlo
posteriormente tambin influye en la seleccin. Elija su opcin y explique por qu.
"""
def analyze_optimal_k(X):
    """
    Analyze optimal number of clusters using elbow method and silhouette analysis

    Parameters
    ----------
    X : ndarray, shape (n_samples, n_features)
        Input data
    """
    print("=== Analyzing Optimal Number of Clusters ===")

    # Elbow method
    print("\n1. Elbow Method:")
    k_values, sse_values = elbow_method(X, max_k=8, verbose=True)

    # Silhouette analysis
    print("\n2. Silhouette Analysis:")
    silhouette_scores = []

    for k in range(2, 9):
        kmeans = ScratchKMeans(n_clusters=k, n_init=10, verbose=False)
        kmeans.fit(X)
        labels = kmeans.predict(X)

        # Calculate silhouette score
        silhouette_vals, silhouette_avg = calculate_silhouette_coefficients(X, labels)
        silhouette_scores.append(silhouette_avg)

        print(f"k={k}: Silhouette Score = {silhouette_avg:.3f}")

    # Plot silhouette scores
    plt.figure(figsize=(10, 6))
    plt.plot(range(2, 9), silhouette_scores, 'bo-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Average Silhouette Score')
    plt.title('Silhouette Analysis')
    plt.grid(True)
    plt.show()

    # Find optimal k
    optimal_k = range(2, 9)[np.argmax(silhouette_scores)]
    print(f"\nOptimal k based on silhouette analysis: {optimal_k}")

    return optimal_k


"""

PROBLEMA 11: Comparacin con grupos conocidos


Region y Channel no se usaron para K-medias, pero son grupos conocidos.
Averige y explique cmo el grupo difiere de estos grupos conocidos.
Primero, visualice cada uno.
"""
def compare_with_known_groups(df, labels):
    """
    Compare K-means clusters with known groups (Channel and Region)

    Parameters
    ----------
    df : pandas.DataFrame
        Dataset with Channel and Region columns
    labels : ndarray, shape (n_samples,)
        K-means cluster labels
    """
    print("=== Comparing K-means Clusters with Known Groups ===")

    # Add cluster labels to dataframe
    df_with_clusters = df.copy()
    df_with_clusters['Cluster'] = labels

    # Analyze Channel distribution
    print("\n1. Channel Distribution by Cluster:")
    channel_cross = pd.crosstab(df_with_clusters['Cluster'], df_with_clusters['Channel'])
    print(channel_cross)

    # Analyze Region distribution
    print("\n2. Region Distribution by Cluster:")
    region_cross = pd.crosstab(df_with_clusters['Cluster'], df_with_clusters['Region'])
    print(region_cross)

    # Visualize distributions
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Channel distribution
    channel_cross.plot(kind='bar', ax=axes[0])
    axes[0].set_title('Channel Distribution by Cluster')
    axes[0].set_xlabel('Cluster')
    axes[0].set_ylabel('Count')
    axes[0].legend(['Channel 1 (Horeca)', 'Channel 2 (Retail)'])

    # Region distribution
    region_cross.plot(kind='bar', ax=axes[1])
    axes[1].set_title('Region Distribution by Cluster')
    axes[1].set_xlabel('Cluster')
    axes[1].set_ylabel('Count')
    axes[1].legend(['Region 1 (Lisbon)', 'Region 2 (Oporto)', 'Region 3 (Other)'])

    plt.tight_layout()
    plt.show()

    return df_with_clusters


"""

PROBLEMA 12: Informacin til para mayoristas

Encuentre informacin que sea til para el mayorista y renala en una oracin.
"""
def analyze_cluster_characteristics(df_with_clusters):
    """
    Analyze cluster characteristics to provide useful insights for wholesalers

    Parameters
    ----------
    df_with_clusters : pandas.DataFrame
        Dataset with cluster labels
    """
    print("=== Cluster Characteristics Analysis ===")

    # Calculate mean spending by cluster
    spending_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']

    cluster_stats = df_with_clusters.groupby('Cluster')[spending_cols].mean()

    print("\nAverage Spending by Cluster:")
    print(cluster_stats.round(2))

    # Analyze cluster sizes
    cluster_sizes = df_with_clusters['Cluster'].value_counts().sort_index()
    print(f"\nCluster Sizes: {dict(cluster_sizes)}")

    # Find most profitable clusters
    cluster_stats['Total_Spending'] = cluster_stats.sum(axis=1)
    cluster_stats['Cluster_Size'] = cluster_sizes

    print("\nCluster Summary:")
    for cluster_id in cluster_stats.index:
        total_spending = cluster_stats.loc[cluster_id, 'Total_Spending']
        size = cluster_stats.loc[cluster_id, 'Cluster_Size']
        avg_spending_per_customer = total_spending / size

        print(f"Cluster {cluster_id}: {size} customers, "
              f"Total spending: {total_spending:.0f}, "
              f"Avg per customer: {avg_spending_per_customer:.0f}")

    # Visualize cluster characteristics
    plt.figure(figsize=(15, 10))

    # Spending patterns
    plt.subplot(2, 2, 1)
    cluster_stats[spending_cols].plot(kind='bar', ax=plt.gca())
    plt.title('Average Spending by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Average Spending')
    plt.xticks(rotation=45)

    # Channel distribution
    plt.subplot(2, 2, 2)
    channel_dist = df_with_clusters.groupby('Cluster')['Channel'].value_counts().unstack()
    channel_dist.plot(kind='bar', ax=plt.gca())
    plt.title('Channel Distribution by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Count')

    # Region distribution
    plt.subplot(2, 2, 3)
    region_dist = df_with_clusters.groupby('Cluster')['Region'].value_counts().unstack()
    region_dist.plot(kind='bar', ax=plt.gca())
    plt.title('Region Distribution by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Count')

    # Cluster sizes
    plt.subplot(2, 2, 4)
    cluster_sizes.plot(kind='bar', ax=plt.gca())
    plt.title('Cluster Sizes')
    plt.xlabel('Cluster')
    plt.ylabel('Number of Customers')

    plt.tight_layout()
    plt.show()

    return cluster_stats


"""

PROBLEMA 13: (Tarea avanzada) Investigacin de otros mtodos

Adems de K-means tratado esta vez, DBSCAN es un mtodo bien conocido para la
agrupacin no jerrquica.
Adems de PCA, t-SNE y LLE son mtodos bien conocidos para la reduccin de dimensionalidad.

Ventajas y desventajas de DBSCAN
Ventajas y desventajas de t-SNE y LLE
Por favor describa acerca de.
"""
def analyze_other_methods():
    """
    Analyze advantages and disadvantages of other clustering and dimensionality reduction methods
    """
    print("=== Analysis of Other Methods ===")

    print("\n1. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)")
    print("Ventajas:")
    print("- No requiere especificar el nmero de clusters a priori")
    print("- Puede encontrar clusters de forma arbitraria")
    print("- Detecta outliers automticamente")
    print("- Robusto a clusters de diferentes tamaos")
    print("- Funciona bien con datos no esfricos")

    print("\nDesventajas:")
    print("- Sensible a los parmetros eps y min_samples")
    print("- Dificultad para encontrar clusters de diferentes densidades")
    print("- No funciona bien con datos de alta dimensionalidad")
    print("- Computacionalmente costoso para datasets grandes")

    print("\n2. t-SNE (t-Distributed Stochastic Neighbor Embedding)")
    print("Ventajas:")
    print("- Excelente para visualizacin de datos de alta dimensionalidad")
    print("- Preserva la estructura local de los datos")
    print("- Muy efectivo para datasets pequeos a medianos")
    print("- Produce visualizaciones muy claras")

    print("\nDesventajas:")
    print("- No preserva distancias globales")
    print("- Computacionalmente costoso")
    print("- Los resultados pueden variar entre ejecuciones")
    print("- No es adecuado para datasets muy grandes")

    print("\n3. LLE (Locally Linear Embedding)")
    print("Ventajas:")
    print("- Preserva la estructura local de los datos")
    print("- No requiere parmetros de escala")
    print("- Funciona bien con datos no lineales")
    print("- Relativamente eficiente computacionalmente")

    print("\nDesventajas:")
    print("- Sensible a la seleccin del nmero de vecinos")
    print("- Puede tener problemas con datos ruidosos")
    print("- No preserva distancias globales")
    print("- Puede colapsar en datasets pequeos")


"""

PROBLEMA 14: (Tarea avanzada) Uso de t-SNE y DBSCAN


Utilice t-SNE en lugar de PCA para reducir las dimensiones de los datos de clientes mayoristas.
Adems, utilice DBSCAN en lugar de K-means para la agrupacin.

Comparemos el resultado de la agrupacin con el de PCA + kmeans.
Utilice scikit-learn para la implementacin.
"""
def compare_tsne_dbscan_with_pca_kmeans(X, df):
    """
    Compare t-SNE + DBSCAN with PCA + K-means

    Parameters
    ----------
    X : ndarray, shape (n_samples, n_features)
        Input data
    df : pandas.DataFrame
        Original dataset
    """
    print("=== Comparing t-SNE + DBSCAN with PCA + K-means ===")

    # PCA + K-means
    print("\n1. PCA + K-means:")
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    kmeans = ScratchKMeans(n_clusters=4, n_init=10, verbose=False)
    kmeans.fit(X_pca)
    labels_kmeans = kmeans.predict(X_pca)

    print(f"PCA explained variance ratio: {pca.explained_variance_ratio_}")
    print(f"K-means SSE: {kmeans.get_sse():.2f}")

    # t-SNE + DBSCAN
    print("\n2. t-SNE + DBSCAN:")
    tsne = TSNE(n_components=2, random_state=42, perplexity=30)
    X_tsne = tsne.fit_transform(X)

    dbscan = DBSCAN(eps=0.5, min_samples=5)
    labels_dbscan = dbscan.fit_predict(X_tsne)

    n_clusters_dbscan = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
    n_noise = list(labels_dbscan).count(-1)

    print(f"Number of clusters found by DBSCAN: {n_clusters_dbscan}")
    print(f"Number of noise points: {n_noise}")

    # Visualize results
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))

    # PCA + K-means
    scatter1 = axes[0, 0].scatter(X_pca[:, 0], X_pca[:, 1], c=labels_kmeans, cmap='viridis')
    axes[0, 0].set_title('PCA + K-means')
    axes[0, 0].set_xlabel('PC1')
    axes[0, 0].set_ylabel('PC2')
    plt.colorbar(scatter1, ax=axes[0, 0])

    # t-SNE + DBSCAN
    scatter2 = axes[0, 1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels_dbscan, cmap='viridis')
    axes[0, 1].set_title('t-SNE + DBSCAN')
    axes[0, 1].set_xlabel('t-SNE 1')
    axes[0, 1].set_ylabel('t-SNE 2')
    plt.colorbar(scatter2, ax=axes[0, 1])

    # Compare with known groups - Channel
    scatter3 = axes[1, 0].scatter(X_pca[:, 0], X_pca[:, 1], c=df['Channel'], cmap='viridis')
    axes[1, 0].set_title('PCA + Known Groups (Channel)')
    axes[1, 0].set_xlabel('PC1')
    axes[1, 0].set_ylabel('PC2')
    plt.colorbar(scatter3, ax=axes[1, 0])

    # Compare with known groups - Region
    scatter4 = axes[1, 1].scatter(X_pca[:, 0], X_pca[:, 1], c=df['Region'], cmap='viridis')
    axes[1, 1].set_title('PCA + Known Groups (Region)')
    axes[1, 1].set_xlabel('PC1')
    axes[1, 1].set_ylabel('PC2')
    plt.colorbar(scatter4, ax=axes[1, 1])

    plt.tight_layout()
    plt.show()

    # Compare cluster characteristics
    print("\n3. Cluster Characteristics Comparison:")

    # K-means clusters
    df_kmeans = df.copy()
    df_kmeans['Cluster'] = labels_kmeans
    kmeans_stats = df_kmeans.groupby('Cluster')[['Channel', 'Region']].agg(['mean', 'count'])

    # DBSCAN clusters
    df_dbscan = df.copy()
    df_dbscan['Cluster'] = labels_dbscan
    dbscan_stats = df_dbscan.groupby('Cluster')[['Channel', 'Region']].agg(['mean', 'count'])

    print("\nK-means Cluster Statistics:")
    print(kmeans_stats)

    print("\nDBSCAN Cluster Statistics:")
    print(dbscan_stats)

    return X_pca, X_tsne, labels_kmeans, labels_dbscan


"""

FUNCIN PRINCIPAL - EJECUTA TODOS LOS PROBLEMAS

"""
def main():
    """Main function to run all K-means examples and analyses"""
    print(" K-means Implementation from Scratch ")
    print("=" * 50)

    # Test with artificial dataset
    print("\n=== Testing with Artificial Dataset ===")
    X_artificial = create_sample_dataset()

    # Test basic K-means
    kmeans = ScratchKMeans(n_clusters=4, n_init=5, verbose=True)
    kmeans.fit(X_artificial)
    labels = kmeans.predict(X_artificial)

    visualize_clusters(X_artificial, labels, kmeans.centroids, "K-means on Artificial Data")

    # Elbow method
    print("\n=== Elbow Method Analysis ===")
    elbow_method(X_artificial, max_k=8, verbose=True)

    # Silhouette analysis
    print("\n=== Silhouette Analysis ===")
    optimal_k = analyze_optimal_k(X_artificial)

    # Test with optimal k
    kmeans_optimal = ScratchKMeans(n_clusters=optimal_k, n_init=10, verbose=True)
    kmeans_optimal.fit(X_artificial)
    labels_optimal = kmeans_optimal.predict(X_artificial)

    # Calculate silhouette coefficients
    silhouette_vals, silhouette_avg = calculate_silhouette_coefficients(X_artificial, labels_optimal)
    plot_silhouette_diagram(X_artificial, labels_optimal, silhouette_vals, silhouette_avg)

    # Wholesale customers analysis
    print("\n=== Wholesale Customers Analysis ===")
    df = load_wholesale_data()

    # Prepare data for clustering (exclude Channel and Region)
    spending_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']
    X_wholesale = df[spending_cols].values

    # Standardize the data
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    X_wholesale_scaled = scaler.fit_transform(X_wholesale)

    # Apply PCA for visualization
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_wholesale_scaled)

    print(f"PCA explained variance ratio: {pca.explained_variance_ratio_}")

    # Find optimal k for wholesale data
    optimal_k_wholesale = analyze_optimal_k(X_pca)

    # Apply K-means with optimal k
    kmeans_wholesale = ScratchKMeans(n_clusters=optimal_k_wholesale, n_init=10, verbose=True)
    kmeans_wholesale.fit(X_pca)
    labels_wholesale = kmeans_wholesale.predict(X_pca)

    # Compare with known groups
    df_with_clusters = compare_with_known_groups(df, labels_wholesale)

    # Analyze cluster characteristics
    cluster_stats = analyze_cluster_characteristics(df_with_clusters)

    # Analyze other methods
    analyze_other_methods()

    # Compare t-SNE + DBSCAN with PCA + K-means
    compare_tsne_dbscan_with_pca_kmeans(X_wholesale_scaled, df)

    print("\n All K-means analyses completed successfully! ")


if __name__ == "__main__":
    main()
```
