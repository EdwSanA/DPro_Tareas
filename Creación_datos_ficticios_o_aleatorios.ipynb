```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Problema 1 ----------
# Semilla para reproducibilidad
np.random.seed(0)

# Distribucin normal bidimensional
mean_1 = [-3, 0]
cov_1  = [[1.0, 0.8],
          [0.8, 1.0]]

# 500 puntos con media (-3,0) y matriz de covarianza dada
data_0 = np.random.multivariate_normal(mean_1, cov_1, 500)
print("Forma de data_0:", data_0.shape)

# ---------- Problema 2 ----------
plt.figure(figsize=(8, 6))
plt.scatter(data_0[:, 0], data_0[:, 1], alpha=0.8)
plt.title("Diagrama de dispersin (Problema 1)")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()

# ---------- Problema 3 ----------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
# Usamos el mismo rango y nmero de bins para comparar mejor
hist_range = (-6, 3)
ax1.hist(data_0[:, 0], bins=50, range=hist_range, rwidth=0.9)
ax1.set_title("Histograma de x1")
ax1.set_xlabel("x1")
ax1.set_ylabel("Frecuencia")
ax1.set_xlim(hist_range)

ax2.hist(data_0[:, 1], bins=50, range=hist_range, rwidth=0.9)
ax2.set_title("Histograma de x2")
ax2.set_xlabel("x2")
ax2.set_ylabel("Frecuencia")
ax2.set_xlim(hist_range)

plt.tight_layout()
plt.show()

# ---------- Problema 4 ----------
mean_2 = [0, -3]
cov_2  = [[1.0, 0.8],
          [0.8, 1.0]]

data_1 = np.random.multivariate_normal(mean_2, cov_2, 500)

plt.figure(figsize=(8, 6))
plt.scatter(data_0[:, 0], data_0[:, 1], label='0', alpha=0.8, c='tab:blue')
plt.scatter(data_1[:, 0], data_1[:, 1], label='1', alpha=0.8, c='tab:orange')
plt.title("Diagrama de dispersin con ambos conjuntos")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()

# ---------- Problema 5 ----------
combined_data = np.concatenate([data_0, data_1], axis=0)
print("Forma de los datos combinados:", combined_data.shape)

# ---------- Problema 6 ----------
# Etiquetas como enteros
labels_0 = np.zeros((500, 1), dtype=int)
labels_1 = np.ones((500, 1), dtype=int)

data_with_labels_0 = np.concatenate([data_0, labels_0], axis=1)
data_with_labels_1 = np.concatenate([data_1, labels_1], axis=1)

final_dataset = np.concatenate([data_with_labels_0, data_with_labels_1], axis=0)

print("\nPrimeros 5 registros del dataset final:\n", final_dataset[:5])
print("\nltimos 5 registros del dataset final:\n", final_dataset[-5:])
print("\nForma final (1000, 3):", final_dataset.shape)
```
