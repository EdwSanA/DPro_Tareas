```python
# AN√ÅLISIS DE CR√âDITO


# üîπ Librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilo
plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")
pd.set_option("display.max_columns", 50)

# =======================================================
# [Problema 1] Comprender el contenido de la competencia
# =======================================================
print(" PROBLEMA 1 - Competencia Home Credit\n")

# Respuesta textual (puedes adaptarlo si lo entregas como informe)
print("""
- Home Credit es una empresa internacional de pr√©stamos al consumo que ofrece microcr√©ditos y financiamiento accesible.
- En esta competencia, se busca predecir la probabilidad de incumplimiento de cr√©dito de los clientes.
- El beneficio para la empresa es poder reducir riesgos financieros, mejorar sus modelos de scoring y ofrecer pr√©stamos m√°s justos y accesibles.
""")


# ====================================================
# [Problema 2] Comprender la descripci√≥n general de los datos
# ====================================================
print("\n PROBLEMA 2 - Exploraci√≥n inicial\n")

# Cargar el dataset (ya subido en Colab)
df = pd.read_csv("application_train.csv")

print("üîπ Vista inicial del dataset:\n")
display(df.head())

print("\nüîπ Informaci√≥n general del dataset:")
print(df.info())

print("\nüîπ Estad√≠sticas descriptivas:")
display(df.describe())

# Valores faltantes
missing = df.isnull().mean().sort_values(ascending=False) * 100
missing = missing[missing > 0]
print("\nüîπ Porcentaje de valores faltantes por columna (top 10):")
display(missing.head(10))

# Proporciones de clase objetivo
plt.figure(figsize=(5,4))
df['TARGET'].value_counts(normalize=True).plot(
    kind='bar', color=['#4CAF50','#F44336'], alpha=0.8
)
plt.title("Proporci√≥n de clases (0 = No incumple, 1 = Incumple)")
plt.xticks(rotation=0)
plt.show()

# ===========================================================
# [Problema 3] Configuraci√≥n de asignaci√≥n (Planteo de retos)
# ===========================================================

print("\n PROBLEMA 3 - Preguntas de investigaci√≥n\n")

print("""
Algunas preguntas que podemos investigar son:

1. ¬øC√≥mo influye el ingreso anual en la probabilidad de incumplimiento?
2. ¬øLa edad del solicitante est√° relacionada con el riesgo de cr√©dito?
3. ¬øExisten diferencias por g√©nero en la tasa de incumplimiento?
4. ¬øEl tipo de contrato de trabajo afecta la probabilidad de impago?
5. ¬øLas personas con propiedades (carro o apartamento) incumplen menos?
6. ¬øLa relaci√≥n entre el monto del cr√©dito solicitado y el ingreso anual predice el riesgo?
7. ¬øLa duraci√≥n del historial crediticio tiene efecto en el incumplimiento?
8. ¬øEl nivel educativo o el estado civil influyen en la probabilidad de impago?
""")



# [Problema 4] Exploraci√≥n de datos (EDA)

print("\n  PROBLEMA 4 - EDA con gr√°ficos\n")

# 1. Distribuci√≥n de ingresos por clase TARGET
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="TARGET", y="AMT_INCOME_TOTAL")
plt.ylim(0, 200000)  # limitar para ver mejor
plt.title("Ingreso anual por clase objetivo")
plt.show()

# 2. Edad de los solicitantes (DAYS_BIRTH es negativo -> convertir a positivo)
df["AGE"] = -df["DAYS_BIRTH"] / 365
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="AGE", hue="TARGET", bins=30, kde=True, element="step")
plt.title("Distribuci√≥n de edad por clase objetivo")
plt.show()

# 3. G√©nero vs incumplimiento
plt.figure(figsize=(5,4))
sns.barplot(x="CODE_GENDER", y="TARGET", data=df)
plt.title("Tasa de incumplimiento por g√©nero")
plt.show()

# 4. Tipo de contrato de trabajo
plt.figure(figsize=(10,5))
sns.barplot(x="NAME_CONTRACT_TYPE", y="TARGET", data=df)
plt.title("Incumplimiento seg√∫n tipo de contrato")
plt.show()

# 5. Propiedad de auto vs incumplimiento
plt.figure(figsize=(5,4))
sns.barplot(x="FLAG_OWN_CAR", y="TARGET", data=df)
plt.title("Incumplimiento seg√∫n propiedad de auto")
plt.show()

# 6) Relaci√≥n cr√©dito/ingreso y riesgo =====
df['CREDIT_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']

print("\nResumen de CREDIT_INCOME_RATIO por TARGET:")
display(
    df.groupby('TARGET')['CREDIT_INCOME_RATIO']
      .describe()[['count','mean','std','min','25%','50%','75%','max']]
)

plt.figure(figsize=(9,5))
sns.boxplot(data=df, x='TARGET', y='CREDIT_INCOME_RATIO')
plt.ylim(0, 6)  # recorte para ver mejor (evita outliers extremos)
plt.title("Relaci√≥n AMT_CREDIT / AMT_INCOME_TOTAL por clase TARGET")
plt.xlabel("TARGET (0 = no incumple, 1 = incumple)")
plt.ylabel("CREDIT_INCOME_RATIO")
plt.show()

# Versi√≥n binned para ver tendencias
df['CIR_BIN'] = pd.qcut(df['CREDIT_INCOME_RATIO'], q=10, duplicates='drop')
ratio_by_bin = df.groupby('CIR_BIN')['TARGET'].mean()
plt.figure(figsize=(10,4))
ratio_by_bin.plot(kind='bar')
plt.title("Tasa de incumplimiento por decil de CREDIT_INCOME_RATIO")
plt.ylabel("Tasa media de TARGET")
plt.xticks(rotation=45)
plt.show()

# 7) A√±os empleados (proxy de estabilidad/historial) y riesgo
# Limpieza: 365243 significa 'desconocido' -> NaN
emp = df['DAYS_EMPLOYED'].replace(365243, np.nan)
df['YEARS_EMPLOYED'] = (-emp) / 365.0  # convertir a a√±os

print("\nResumen de YEARS_EMPLOYED por TARGET:")
display(
    df.groupby('TARGET')['YEARS_EMPLOYED']
      .describe()[['count','mean','std','min','25%','50%','75%','max']]
)

plt.figure(figsize=(9,5))
sns.kdeplot(data=df, x='YEARS_EMPLOYED', hue='TARGET', common_norm=False, fill=True, alpha=0.3)
plt.xlim(0, 50)  # rango razonable para visualizar
plt.title("Distribuci√≥n de a√±os empleados por clase TARGET")
plt.xlabel("A√±os empleados")
plt.show()

# Bins para ver tasa de incumplimiento por tramo de antig√ºedad
df['EMP_BIN'] = pd.cut(df['YEARS_EMPLOYED'], bins=[0,1,3,5,10,20,50])
emp_risk = df.groupby('EMP_BIN')['TARGET'].mean()
plt.figure(figsize=(10,4))
emp_risk.plot(kind='bar')
plt.title("Tasa de incumplimiento por tramo de a√±os empleados")
plt.ylabel("Tasa media de TARGET")
plt.xticks(rotation=0)
plt.show()

# 8) Educaci√≥n y estado civil vs riesgo
cats = ['NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS']

for col in cats:
    plt.figure(figsize=(9,5))
    # Tasa de incumplimiento media por categor√≠a (ordenada)
    order = df.groupby(col)['TARGET'].mean().sort_values(ascending=False).index
    sns.barplot(data=df, x=col, y='TARGET', order=order)
    plt.title(f"Tasa de incumplimiento por {col}")
    plt.ylabel("Tasa media de TARGET")
    plt.xlabel(col)
    plt.xticks(rotation=30, ha='right')
    plt.show()

# Tabla resumen combinada (√∫til para el informe)
summary = (
    df.groupby(['NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS'])['TARGET']
      .mean()
      .reset_index()
      .sort_values('TARGET', ascending=False)
)
print("\nTasa de incumplimiento por combinaci√≥n EDUCACI√ìN x ESTADO CIVIL (top 12):")
display(summary.head(12))

```
