Problema 01: Función lineal

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-50, 50.1, 0.1)  # 50.1 to include 50

# Calculate y = 1/2x + 1
y = 1/2 * x + 1

print("elements of x:", x)
print("elements of y:", y)
```

Problema 02: combinado array

```python
#Este código combina los arrays x e y columna por columna (como si creara una tabla de pares ordenados
#(𝑥,𝑦)
#(x,y)) y luego imprime su forma y contenido.

xy = np.column_stack((x, y))

print("Shape of combined array:", xy.shape)
print("Combined array elements:", xy)
```

problema 03: encontrar la pendiente

```python
#Este código calcula la pendiente (gradiente) entre pares consecutivos de puntos de la función
# 𝑦=𝑓(𝑥)
# y=f(x), usando diferencias finitas.

dy = np.diff(y)  # Differences between adjacent y values
dx = np.diff(x)  # Differences between adjacent x values
gradient = dy / dx

print("Gradient values:", gradient)
print("Length of gradient array:", len(gradient))
```

Problema 05:"Funcionalización en Python" o también "Convertir en función en Python",
1. 𝑦=𝑥2, (−50≤𝑥≤50)

 𝑦=2𝑥2+2𝑥,(−50≤𝑥≤50) y=2x 2 +2x,(−50≤x≤50)

```python
# Esta función:
# Recibe una función matemática y un rango de valores x.
# Calcula y = f(x).
# Obtiene el gradiente como dy/dx.
# Devuelve los pares (x,y) y el gradiente.

import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.

    Parameters
    ----------
    function : function
        A function that returns ndarray y when given ndarray x.
    x_range : tuple
        Range like (start, stop, step) for np.arange().

    Returns
    -------
    array_xy : ndarray, shape(n, 2)
        Combined x and y values.
    gradient : ndarray, shape(n - 1)
        Gradient (dy/dx) between adjacent points.
    """
    x = np.arange(*x_range)  # create x values
    y = function(x)          # calculate y values using the provided function
    dy = np.diff(y)
    dx = np.diff(x)
    gradient = dy / dx
    array_xy = np.column_stack((x, y))
    return array_xy, gradient
```

```python
# Este código:
# Define tres funciones matemáticas.
# Usa compute_gradient para calcular los gradientes de cada una en el rango especificado.

# compute required functions
def function1(x):
    return x ** 2

def function2(x):
    return 2 * x ** 2 + 2 * x

def function3(x):
    return np.sin(x / 12)

# compute gradients for all 3 functions
# For y = x^2
xy1, grad1 = compute_gradient(function1)

# For y = 2x^2 + 2x
xy2, grad2 = compute_gradient(function2)

# For y = sin(x/12), x from 0 to 50
xy3, grad3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))  # as per given range
```

```python
# Draw graphs
def function_gradient_plot(xy, gradient, title_prefix):
    x = xy[:, 0]
    y = xy[:, 1]
    x_grad = x[:-1]

    plt.figure(figsize=(12, 5))

    # Plot 1: Function
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label=f'{title_prefix} Function')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{title_prefix} Function')
    plt.grid(True)
    plt.legend()

    # Plot 2: Gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_grad, gradient, label=f'{title_prefix} Gradient', color='orange')
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.title(f'{title_prefix} Gradient')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

# Plot all three
function_gradient_plot(xy1, grad1, "y = x^2")
function_gradient_plot(xy2, grad2, "y = 2x^2 + 2x")
function_gradient_plot(xy3, grad3, "y = sin(x/12)")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el gradiente
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = function(x)
    dy = np.diff(y)
    dx = np.diff(x)
    gradient = dy / dx
    array_xy = np.column_stack((x, y))
    return array_xy, gradient

# Definir las funciones matemáticas
def function1(x):
    return x ** 2

def function2(x):
    return 2 * x ** 2 + 2 * x

def function3(x):
    return np.sin(x / 12)

# Graficar función y gradiente
def function_gradient_plot(xy, gradient, title_prefix):
    x = xy[:, 0]
    y = xy[:, 1]
    x_grad = x[:-1]

    plt.figure(figsize=(12, 5))

    # Gráfico de la función
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label=f'{title_prefix} Function')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{title_prefix} Function')
    plt.grid(True)
    plt.legend()

    # Gráfico del gradiente
    plt.subplot(1, 2, 2)
    plt.plot(x_grad, gradient, label=f'{title_prefix} Gradient', color='orange')
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.title(f'{title_prefix} Gradient')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

# Calcular gradientes
xy1, grad1 = compute_gradient(function1)
xy2, grad2 = compute_gradient(function2)
xy3, grad3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))

# Mostrar las gráficas
function_gradient_plot(xy1, grad1, "y = x^2")
function_gradient_plot(xy2, grad2, "y = 2x^2 + 2x")
function_gradient_plot(xy3, grad3, "y = sin(x/12)")
```

[Problema 6] Encuentre el valor mínimo

```python
#Este bloque:
#Evalúa el punto mínimo de cada función y el comportamiento del gradiente justo antes y después.
#También incluye el caso especial de y = sin(x/12) con rango ajustado.

def analyze_minimum(xy, gradient, func_label):
    y = xy[:, 1]
    x = xy[:, 0]

    min_index = np.argmin(y)
    min_y = y[min_index]
    min_x = x[min_index]

    print(f"\n--- {func_label} ---")
    print(f"Minimum y = {min_y:.4f} at x = {min_x:.4f}")

    # Show slope before and after, if they exist
    if 1 <= min_index < len(gradient):
        print(f"Slope before minimum: {gradient[min_index - 1]:.4f}")
        print(f"Slope after minimum: {gradient[min_index]:.4f}")
    else:
        print("Minimum is at the edge of the domain, cannot compute both slopes.")

# Function 1: y = x^2
xy1, grad1 = compute_gradient(function1)
analyze_minimum(xy1, grad1, "y = x^2")

# Function 2: y = 2x^2 + 2x
xy2, grad2 = compute_gradient(function2)
analyze_minimum(xy2, grad2, "y = 2x^2 + 2x")

# Function 3: y = sin(x/12), x in [0, 50]
xy3, grad3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))
analyze_minimum(xy3, grad3, "y = sin(x/12)")
```
