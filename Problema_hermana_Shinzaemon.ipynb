**Problema con la hermana Shinzaemon**

Hay una anécdota sobre Soro Shinzaemon similar a la historia de doblar un trozo de papel para alcanzar la luna en el problema "Intentemos usar Python". Procederemos a programar basándonos en esto.

Cuando Hideyoshi le ofreció una recompensa, le preguntaron a Shinzaemon qué quería recibir, y pidió un grano de arroz hoy, dos granos al día siguiente, cuatro granos al otro día, y así sucesivamente, duplicando la cantidad de arroz cada día durante 100 días. Pensando que un grano de arroz no sería gran cosa, Hideyoshi aceptó de inmediato, pero a mitad de camino se dio cuenta de que duplicar la cantidad de arroz cada día resultaría en una cantidad enorme al cabo de 100 días, así que cambió la recompensa.

**Problema 1 Número de granos de arroz en el día 100**

Crea un código para calcular cuántos granos de arroz recibes en total el día 100 y genera el resultado. Además, usa una gráfica de líneas para representar la variación en la cantidad de granos de arroz que recibes ese día y la cantidad total de granos de arroz que recibes ese día, del día 1 al día 100. (El eje vertical representa la cantidad de arroz y el eje horizontal, la cantidad de días).

```python
import matplotlib.pyplot as plt

# Número de días
days = 100

# Listas para almacenar datos
granos_dia = []      # granos recibidos en cada día
granos_acumulados = []  # total acumulado
total = 0

# Cálculo día por día
for d in range(1, days + 1):
    granos_hoy = 2 ** (d - 1)          # se duplica cada día
    total += granos_hoy
    granos_dia.append(granos_hoy)
    granos_acumulados.append(total)

# Resultado del día 100
print("Granos en el día 100:", granos_dia[-1])
print("Total acumulado al día 100:", granos_acumulados[-1])

# --- Gráfica ---
plt.figure(figsize=(10,6))
plt.title("Cantidad de granos de arroz hasta el día 100")
plt.xlabel("Día")
plt.ylabel("Número de granos")
plt.plot(range(1, days + 1), granos_dia, label="Granos recibidos por día", color="blue")
plt.plot(range(1, days + 1), granos_acumulados, label="Total acumulado", color="green")
plt.yscale('log')  # escala logarítmica para apreciar el crecimiento
plt.grid(True, which="both", ls="--", alpha=0.6)
plt.legend()
plt.show()
```

**Problema 2 Número de granos de arroz fuera del día 100**

El número de días en esta anécdota es ambiguo, como 30 u 81 días. Creemos una función que tome días como argumento para que pueda manejar cualquier historia. Después de crear la función, indique cuántos granos de arroz recibirá en los días que tome el curso. Genere la gráfica de líneas y el problema 1.


```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    Calcula la cantidad de granos de arroz en la historia de Sorori Shinzaemon.

    Parámetros
    ----------
    n_days : int, opcional
        Número de días de la historia (por defecto 100).

    Retorna
    -------
    list_n_grains : list[int]
        Lista con la cantidad de granos recibidos cada día
        (la longitud de la lista es igual a n_days).
    list_total_grains : list[int]
        Lista con el total acumulado de granos hasta cada día
        (la longitud de la lista es igual a n_days).
    """
    grain = 1                   # granos que se entregan el primer día
    total = 0                   # total acumulado
    list_n_grains = []          # granos que se reciben cada día
    list_total_grains = []      # total acumulado hasta cada día

    for _ in range(n_days):
        list_n_grains.append(grain)
        total += grain
        list_total_grains.append(total)
        grain *= 2              # se duplica cada día

    return list_n_grains, list_total_grains


# ======== Ejemplo de uso ======== #
# Cambia n_days para cualquier historia (30, 81, 100, etc.)
dias = 81
granos_dia, granos_totales = compute_sorori_shinzaemon(n_days=dias)

print(f"Granos recibidos el día {dias}: {granos_dia[-1]}")
print(f"Total acumulado hasta el día {dias}: {granos_totales[-1]}")

# ======== Gráfica ======== #
plt.figure(figsize=(10,6))
plt.title(f"Granos de arroz durante {dias} días")
plt.xlabel("Día")
plt.ylabel("Número de granos")
plt.plot(range(1, dias + 1), granos_dia, label="Granos por día", color="blue")
plt.plot(range(1, dias + 1), granos_totales, label="Total acumulado", color="green")
plt.yscale('log')  # Escala logarítmica para apreciar el rango completo
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

**【Problema 3] ¿Cuántas personas pueden vivir durante cuántos días con el arroz que conseguimos?**
Crea una función para calcular cuántas personas pueden vivir durante cuántos días con este arroz. Indica cuántos días pueden vivir las personas de esta clase con los granos de arroz que reciben durante el curso.

La entrada mínima para la función es el número de granos de arroz (int) y el número de personas (int) , y la salida es el número de días de vida (int) . Asegúrese de que el nombre y los comentarios de la función sean fáciles de entender.

No se proporciona aquí la cantidad de granos de arroz necesarios al día. Por favor, investigue por su cuenta y explique la fórmula. Deje un comentario sobre cómo realizó su investigación y en qué se basó.

```python
import math

def compute_sorori_shinzaemon(n_days=100):
    """
    Devuelve:
      - grains_per_day: granos recibidos cada día (longitud = n_days)
      - grains_cumulative: total acumulado hasta cada día (longitud = n_days)
    """
    grain = 1
    total = 0
    grains_per_day = []
    grains_cumulative = []
    for _ in range(n_days):
        grains_per_day.append(grain)
        total += grain
        grains_cumulative.append(total)
        grain *= 2
    return grains_per_day, grains_cumulative


def days_supported_with_rice(total_grains: int,
                             n_people: int,
                             kcal_per_person_per_day: float = 2100.0,
                             kcal_per_100g_rice_raw: float = 365.0,
                             thousand_grain_weight_g: float = 25.0) -> int:
    """
    Calcula cuántos días puede vivir un grupo únicamente con arroz.

    Parámetros mínimos:
      - total_grains (int): número total de granos disponibles.
      - n_people (int): número de personas.

    Parámetros opcionales (puedes ajustarlos si cambias los supuestos):
      - kcal_per_person_per_day: kcal necesarias por persona y día (por defecto 2100).
      - kcal_per_100g_rice_raw: kcal por 100 g de arroz CRUDO (por defecto 365).
      - thousand_grain_weight_g: peso de 1000 granos en gramos (por defecto 25 g).

    Retorna:
      - días (int): número entero de días que alcanza el arroz para todo el grupo.
    """
    if n_people <= 0:
        raise ValueError("n_people debe ser > 0")
    if total_grains < 0:
        raise ValueError("total_grains no puede ser negativo")

    kcal_per_g = kcal_per_100g_rice_raw / 100.0        # ~3.65 kcal/g
    grams_per_person_per_day = kcal_per_person_per_day / kcal_per_g
    grams_per_grain = thousand_grain_weight_g / 1000.0 # g/grano
    grains_per_person_per_day = math.ceil(grams_per_person_per_day / grams_per_grain)

    # días enteros que cubre para todo el grupo
    return int(total_grains // (n_people * grains_per_person_per_day))


# ======== EJEMPLO: usar los granos del curso y el tamaño de tu clase ========

# 1) Define la duración del curso (días) y el tamaño de la clase:
curso_dias = 100        # cámbialo si tu curso dura 30, 81, etc.
n_personas = 30         # <-- pon aquí el número real de tu clase

# 2) Calcula los granos totales obtenidos en el curso (día 1..curso_dias):
_, acumulado = compute_sorori_shinzaemon(curso_dias)
total_granos_curso = acumulado[-1]     # = 2**curso_dias - 1

# 3) ¿Cuántos días puede vivir el grupo solo con ese arroz?
dias_soporte = days_supported_with_rice(total_granos_curso, n_personas)

print(f"Total de granos recibidos en {curso_dias} días: {total_granos_curso}")
print(f"Con {n_personas} personas, el arroz alcanzaría para ~{dias_soporte} días (solo arroz).")
```
